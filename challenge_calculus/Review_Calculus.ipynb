# Bronze medal: Integral of a Polynomial

a) Student A carefully examine his data and proposed a model:

Can you help Student C find the function that demonstrates clicks per hour as a function of time?

c) Use the best model to calculate: at which hour does clicks per hour reach its maximum value? (rounded to two decimal places)

```python
from scipy.optimize import fsolve

# Define Student B's clicks per hour function (Exponential Decay)
def clicks_per_hour_B(x):
    return 1000 * x * np.exp(-x / 8)

# Define the derivative of Student B's function
def derivative_clicks_per_hour_B(x):
    return 1000 * np.exp(-x / 8) * (1 - x / 8)

# Use fsolve to find the critical point where the derivative is 0
critical_point = fsolve(derivative_clicks_per_hour_B, x0=5)[0]

# Calculate the maximum clicks per hour
max_clicks = clicks_per_hour_B(critical_point)

# Print the hour of maximum clicks, rounded to two decimal places
print(f"Maximum clicks per hour occur at: {critical_point:.2f} hours")
```

d) This YouTuber happened to get a double-coupon. This special double-coupon can doubles his profits from all viewings for 5 hours and he can choose the exact time to start. Based on the model we found, can you suggest which hour he should start using this coupon? The time is counted as 0 when his video is published. (rounded to two decimal places)

```python
from scipy.integrate import quad

# Define Student B's clicks per hour function (Exponential Decay)
def clicks_per_hour_B(x):
    return 1000 * x * np.exp(-x / 8)

# Function to calculate total clicks in a 5-hour interval starting from t
def total_clicks_over_interval(t):
    result, _ = quad(clicks_per_hour_B, t, t + 5)
    return result

# Generate time intervals from 0 to 19 (since the last interval will be 19 to 24)
time_intervals = np.linspace(0, 19, 100)

# Calculate total clicks for each interval
total_clicks = [total_clicks_over_interval(t) for t in time_intervals]

# Find the time that maximizes the total clicks
max_total_clicks = np.max(total_clicks)
best_start_time = time_intervals[np.argmax(total_clicks)]

# Print the best start time rounded to two decimal places
print(f"The best time to start using the double-coupon is: {best_start_time:.2f} hours")
```
